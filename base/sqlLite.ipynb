```python
import sqlite3
import pandas as pd
con=sqlite3.connect('data/chinook.db')
tracks=pd.read_sql_query("SELECT * FROM tracks", con)
genres=pd.read_sql_query("SELECT * FROM genres",con)
genres.head()
```

```python
tracks.head()
```

```python
tracks_sub_set=tracks.drop_duplicates('GenreId',keep='first')
tracks_sub_set.head()
```

```python
genre_track=genres.merge(tracks_sub_set[['TrackId','GenreId','Milliseconds']],on='GenreId',how='left')
genre_track.head()
```
