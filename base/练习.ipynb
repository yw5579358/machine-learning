```python
import numpy as np
np.linspace(3.07,50.81,11)
tips = pd.read_csv('data/tips.csv')
```

```python
fig=plt.figure(figsize=(16,8))
```

```python
plt.hist(tips['total_bill'],bins=100)
```

```python
plt.figure(figsize=(16,8))
plt.scatter(tips['total_bill'],tips['tip'])
```

```python
def recode_sex(sex):
    if sex=='Female':
        return 'r'
    else:
        return 'b'
tips['sex_color'] = tips['sex'].apply(recode_sex)
```

```python
plt.figure(figsize=(16,8))
plt.scatter(tips['total_bill'],tips['tip'],c=tips['sex_color'],s=tips['size']*10,alpha=0.5)
tips.head()
```

```python
reviews = pd.read_csv('data/winemag-data_first150k.csv',index_col=0)
reviews.head()
reviews.describe()
```

```python
kwargs=dict(figsize=(16,8),fontsize=11,color = ['b','orange','g','r','purple','brown','pink','gray','cyan','y'])
```

```python
reviews['province'].value_counts().head(10).plot.bar(**kwargs)
```

```python
reviews.points.value_counts().sort_index().plot.area(grid=True)
```

```python
reviews[reviews['price']<100].sample(100).plot.scatter(x='price',y='points',figsize=[16,8],fontsize=11)
```

```python
reviews[reviews['price']<100].plot.hexbin(x='price',y='points',gridsize=20,figsize=[14,8])
```
